```python
df_filter['Fjob'] = df_filter['Fjob'].replace(['teacher' ,'other' ,'services' ,'health' ,'at_home'],[1,2,3,4,5])
df_filter['reason'] = df_filter['reason'].replace(['course' ,'other' ,'home' ,'reputation'],[1,2,3,4])
df_filter['guardian'] = df_filter['guardian'].replace(['mother', 'father' ,'other'],[1,2,3])
```

```python
df_filter.G2 = df_filter.G2.astype(float)
df_filter.G3 = df_filter.G3.astype(float)
df_filter.Fjob = df_filter.Fjob.astype(float)
df_filter = df_filter.replace(np.nan,-1)
```

# La modelación mediante regresión de ambos problemas. 

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
modelo_dfg1 = smf.ols('G1 ~ school+sex+age+address+famsize+Pstatus+Medu+Fedu+Mjob+Fjob+reason+guardian+traveltime+studytime+failures+schoolsup+famsup+paid+activities+nursery+higher+internet+romantic+famrel+freetime+goout+Dalc+Walc+health+absences',df_filter).fit()
modelo_dfg2 = smf.ols('G2 ~ school+sex+age+address+famsize+Pstatus+Medu+Fedu+Mjob+Fjob+reason+guardian+traveltime+studytime+failures+schoolsup+famsup+paid+activities+nursery+higher+internet+romantic+famrel+freetime+goout+Dalc+Walc+health+absences',df_filter).fit()
modelo_dfg3 = smf.ols('G3 ~ school+sex+age+address+famsize+Pstatus+Medu+Fedu+Mjob+Fjob+reason+guardian+traveltime+studytime+failures+schoolsup+famsup+paid+activities+nursery+higher+internet+romantic+famrel+freetime+goout+Dalc+Walc+health+absences',df_filter).fit()
```

```python
modelo_dfg1.summary()
```

la capacidad de este modelo para explicar que ocurre con G1, es cercano al 28%, seria deseable analizar las variables que realmente importan en este modelo

failures,schoolsup,famsup,goout, en un primer vistazo en el 1 periodo de clases, vemos la reelevancia en el resultado de nuestro modelo, se ve que la cantidad de clases reprobadas afecta mucho, si poseen apoyo educacional si poseen apoyo de su familia y tienen salidas con sus amigos, se añade el tiempo de estudio, ya que como hipotesis considero que esto afecta el rendimiento final en cualquier alumno
