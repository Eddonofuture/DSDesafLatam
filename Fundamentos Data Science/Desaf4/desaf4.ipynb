```python
import pandas as pd
```

```python
df = pd.read_csv("flights.csv")
```

```python
passengersMean = df["passengers"].mean()
passengersMean
```

```python
df["underperforming"] = 0
```

```python
for rowname, rowserie in df.iterrows():
    if(rowserie["passengers"] < passengersMean):
        #print(rowserie["passengers"]< passengersMean,rowserie["passengers"], passengersMean)
        df.at[rowname, "underperforming"] = 1
```
