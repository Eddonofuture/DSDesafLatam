3. Estime el puntaje Z


```python
# gle_cgdpc , wdi_exph
```

```python
def getzscore(df,val):
    '''Obtienes el zscore del dataframe'''
    prepoc = df[val].dropna()
    get_mean = prepoc.mean()
    get_std = prepoc.std()
    df['Z_'+val] = (df[val].dropna() - get_mean) / get_std
    df['Z_'+val].dropna()
    return 
```

```python
getzscore(df,'wdi_exph')
```

# 5. Genere un método que devuelva un gráfico comparando dos curvas de densidad

```python
from scipy.stats import norm, gaussian_kde as gkde
```

```python
def graf(df1,df2,variable,log=False,):
    '''Se ingresan 2 dataframe para ser comparados por medio de sus graficos'''
    gdp = df1[variable]
    gdp2 = df2[variable]
    if(log == False):
        plt.subplot(2,1,1)
        plt.title('Primer Data Frame')
        plt.hist(gdp)
        plt.subplot(2,1,2)
        plt.title('Segundo Data Frame')
        plt.hist(gdp2)
    else:
        #modo logaritmo
        plt.subplot(2,1,1)
        plt.title('Primer Data Frame')
        plt.hist(np.log(gdp))
        plt.subplot(2,1,2)
        plt.title('Segundo Data Frame')
        plt.hist(np.log(gdp2))
    return
```

```python
graf(df_sup,df_inf,'imf_pop',True)
```

```python
graf(df_sup,df_inf,'gle_cgdpc',True)
```

```python
graf(df_sup,df_inf,'undp_hdi',False)
```

En los casos de donde los indices son imf_pop y gle_cgdpc es necesari utilizar sus funciones logaritmicas, intentando expresar la normalidad de una mejor forma
