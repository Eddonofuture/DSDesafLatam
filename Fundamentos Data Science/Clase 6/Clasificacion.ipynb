```python
def print_df(df):
    import matplotlib.pyplot as plt
    import seaborn as sns
    '''
    Plot DataFrame en base a sus tipos de variables
    '''
    for n, i in enumerate(df):
        plt.subplot(2, 3, n+1)
        if len(df[i].value_counts()) > 2:
            sns.distplot(df[i])
            plt.title(i)
            plt.xlabel("")
        else:
            sns.countplot(y=df[i])
            plt.title(i)
            plt.xlabel("")
            plt.tight_layout()
    return
```
